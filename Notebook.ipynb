# TASK #4: PERFORM DATA CLEANING

```python
resume_df['resume_text'] = resume_df['resume_text'].apply(lambda x: x.replace('\r',''))
resume_df
```

```python
# download nltk packages
nltk.download('punkt')
```

```python
# download nltk packages
nltk.download("stopwords")
```

```python
# Get additional stopwords from nltk
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
stop_words
```

```python
stop_words.extend(['from', 'subject', 're', 'edu', 'use', 'email', 'com'])
```

```python
# Remove stop words and remove words with 2 or less characters
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 2 and token not in stop_words:
            result.append(token)
    
    return ' '.join(result)
```

```python
# Cleaned text
resume_df['cleaned'] = resume_df['resume_text'].apply(preprocess)
```

# TASK#7: TRAIN NAIVE BAYES CLASSIFIER MODEL

```python
countvectorizer

# countvectorizer was the sparse matrix that has all my training data that all my cleaned up encoded data.
```

```python
X = countvectorizer
```

```python
y = resume_df['class']
```

```python
X.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.naive_bayes import MultinomialNB

NB_classifier = MultinomialNB()
NB_classifier.fit(X_train, y_train)
```

# TASK #8: ASSESS TRAINED MODEL PERFORMANCE


```python
# Predicting the performance on train data
y_predict_train = NB_classifier.predict(X_train)
y_predict_train
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot = True)
```

We can see here, the model did pretty amazing. Actually, the model was able to classify all the different classes correctly.
The two positives, the true negatives were perfect. And we had zero samples that have been misclassified.
Please know that that was on the training data, though.


```python
# Predicting the Test set results
y_predict_test = NB_classifier.predict(X_test)
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True)
```

```python
# classification report
print(classification_report(y_test, y_predict_test))
```

As we can see here, we have been able to correctly classify 18 samples and two samples and misclassify
three plus two or five. And the classification report here is showing that we have achieved around 80% accuracy overall and we have here
the results for precision recall and F one score. And this is simply the harmonic mean between precision
